# Data-driven Approach to decide where to live in London

## 1. Aims, objectives, and background

### 1.1	Introduction

Thousands of people move homes every year in London, 24,970 people moved to a new home in the capital in the first half of the year (2021) compared to 9,440 in the same six months the year before (2020) [1]. There are many factors involved in the decision-making process , one of the most important factors for people moving to a borough in London is the crime rate in the area, Important factor to consider when you decide where to live is the safety of an area [2]. Many people make decisions based on intuition, but this project takes a scientific data-driven approach to make an informed decision about living in a London borough based on the crime rate in the area. Data-driven decision making is defined as the process of using data to inform your decision-making process and validate a course of action before committing to it. [3]

### 1.2	Aims and objectives

For this project the following research questions are explored:
- Is it possible to get access to reliable and consistent data about different crimes in each London borough?
- Using data to answer the following questions:
    1. Can all London boroughs be compared to each other for major crime categories over a period of time?
    2. Can all London boroughs be compared to each other for a specific crime over a period of time?
    3. Since population of each London borough is different so can all London boroughs be compared to each other for a specific crime per person (per capita crime) over a period of time?
    4. How different categories of crimes compare to each other in a specific London borough over a period of time? 
    5. And more importantly, can the trend of any given crime be compared to the trend of another crime in a specific London borough over a period of time?

For this project, the aims are as follows:
1.	Since credibility of organization that provides data is critical for the decision-making process so finding a reliable organization that can be used as data source to provides consistent data. 
2.	Taking into consideration constraints on time and resource, decide on how much data is required to sufficiently explore the above questions.
3.	Collect the data and store in a form which allows for data cleaning and analysis.
4.	Clean and transform the data such that it is suitable to create visualizations.
5.	Use visualisations to carry out exploratory data analysis, resulting in clear decision making.

### 2.3	Limitations and constraints of the data

Data available on London Datastore has the following limitations and constraints:
1.	Most recent borough level crime data available is for last 24 months (From Nov. 2020 to Oct. 2022). Historic data is available for 10 years (From Apr. 2010 to Oct. 2020) separately. 
2.	More granular data such as weekly or daily data is not available.
3.	Data for towns within Borough are not available. Since borough is relatively big so town level crime data within each borough could be more relevant while making decisions to live in an area.
4.	Met. Police is the source of data to London Datastore and it is mentioned on its own website that there are occasions when victims are unable to provide the actual location of a crime. In these instances, the site will not be able to display all the crime reported to the police.[6]
5.	Data on London Datastore is updated in real time by other sources such as Met. Police. This means that there is a delay between crime happening and being updated on London Datastore which has been used as the source of data. Met Police mentioned on their website that "Crime data is extracted from the crime reporting information system (CRIS) every month." [6]
6.  The population data for the year 2021 has been used which is based on projection.    

For meaningful exploratory data analysis of this project, monthly crime data of each borough for last 2 years is sufficient to create visualisations that can be used to find trend for any crime. Other limitations mentioned above are not expected to have significant impact on the data analysis.

### 2.9 Data Extraction

Since data analysis is not focused on a specific borough or a specific crime so global variables have been used for borough and crime, making it easier for user to simply change values of global variables and get analysis for any borough or crime. 

```python
specific_borough = 'Hounslow'
specific_crime = 'Possession of Article with Blade or Point'
#Read CSV file that contains crime figures for each London borough
crime_df = pd.read_csv("Crime.csv")
```

### 2.10 Exploratory Data Analysis

There are two main dataframes used for analysis in later sections
1. crime_df - This data frame contains list of boroughs, categories of crimes in each borough and the number of crimes in each month during two years.
2. population_df - It contains projected data of population in each borough in 2021.

```python
# Remove rows that does not have population data
crime_df = crime_df[(crime_df["LookUp_BoroughName"] != "Aviation Security (SO18)") & (
    crime_df["LookUp_BoroughName"] != "London Heathrow and London City Airports")]
# Read CSV file that contains population figures for each London borough
population_df = pd.read_csv("Population.csv")
# Read population projection figures for year 2021 for each London borough
yearly_population = list(population_df["2021"])
# Validation - since population of any borough cannot be zero so checking data and selecting values if they are greater than 0
yearly_population = list(filter(lambda n: n > 0, yearly_population))
# create condition to select data for only one London Borough, that is, 'Hounslow'
condition = crime_df['LookUp_BoroughName'] == specific_borough
# set condition to get DataFrame
resultDf = (crime_df[condition])
```

#### 2.10.1 Major Crimes in all London boroughs

User needs to see the big picture looking at major crimes happening in all London boroughs before zooming into each borough. To achieve it, a visualization is required that shows all categories of crimes in each and every borough. Since this visualisation is big in size so to make it useful for user, all categories of crimes which have less then 100 crimes in period of two years have been removed. That keeps the focus on categories with significantly more crimes.

A dataframe is created which contains borough names, category of crime and number of crimes as coloums. This dataframe is used to create the visualisation as shown below, making it easier for user to choose a brough.

```python
#Select all the crimes and sum/add for all the months
x_axis = list(crime_df.iloc[:, 3:27].sum(axis='columns'))
y_axis = list(crime_df["LookUp_BoroughName"])
category = list(crime_df["MinorText"])

#create dataframe that contains borough names, crime categories and number of crimes in each category
all_crimes_all_boroughs_df = pd.DataFrame(
    dict(
        borough = y_axis,
        category = category,
        crimes = x_axis

    )
)
# remove categories with crimes less than 100 to make visualization more useful
most_crimes_all_borough_df = all_crimes_all_boroughs_df[all_crimes_all_boroughs_df["crimes"] > 100]

#set font scale for plot
sns.set(font_scale=0.7)
#create facetgrid graph with 2 items in each rows and with 5 as height of each item
g = sns.FacetGrid(most_crimes_all_borough_df, col='borough',
                  sharey=False, col_wrap=2, height=5)
#make layout of graph fit into the available space
g.fig.tight_layout()
#draw bar plot with x-axis representing crimes and y-axis representing categories of crimes
g.map_dataframe(sns.barplot,x='crimes',y='category')
#set lable of x axis as 'No. of Crimes'
g.set_axis_labels('No. of Crimes', fontsize=9)
```



#### 2.10.2 Specific crime in all London boroughs

Since there is no single crime category that is important for everyone so when it comes to choosing borough to live based on crime rate, everyone have their own preference e.g., for a family with teenagers it is important to live in a place where there are less knife crimes but for a family of very young children or old people who stay at home most of the time, it is important to live in a place where there are less home burglaries. This means that the very first step in data analysis is to create a function that takes crime as parameter and returns a list of all London boroughs with crime rate of the given crime as shown below. 

Since population of each borough is different from others so number of crimes is divided by the population of borough to find per capita/person crime. 

```python
###########################################
#Show specific crime in all London boroughs
###########################################

def getBoroughsForSpecificCrime(crime_name):
    #List all london boroughs
    y_axis = crime_df['LookUp_BoroughName'].unique()
    #Create condition to select only crime named 'Possession of Article with Blade or Point'
    condition_knife = crime_df["MinorText"] == crime_name
    #Use condition to get DataFrame containing "Possession of Article with Blade or Point" crime in all London boroughs
    result_knife_df = crime_df[condition_knife]
    #Sum/add for all the months to get total number for each borough, then use it on X-axis of graph
    x_axis = list(result_knife_df.iloc[:, 3:].sum(axis='columns'))

    # create array to store per capita crimes 
    per_capita_crime = []
    # loop through list of crimes in each borough and population of each borough
    for i in range(0,len(x_axis)):
        # divide no. of crime by population for each borough
        per_capita_crime.append(x_axis[i]/yearly_population[i])
    
    #Create new DataFrame so it can be sorted and used for plotting
    specific_crime_all_borough_df = pd.DataFrame(
        dict(
            names = y_axis,
            crimes = per_capita_crime

        )
    )
    #Sort DataFrame
    df_sorted = specific_crime_all_borough_df.sort_values("crimes")
    return df_sorted

def drawBarGraph(dfSorted,graphColor,width,height,title,xlabel,ylabel):
    #Set graph size
    plt.figure(figsize=(width, height))
    #Set sorted DataFrame and other parameters to draw bar graph
    plt.barh("names","crimes", data = dfSorted, color = graphColor)
    #Set title of graph
    plt.title(title, fontsize = 17)
    #Set label of X-axis
    plt.xlabel(xlabel, fontsize = 15)
    #Set label of Y-axis
    plt.ylabel(ylabel, fontsize = 15)

drawBarGraph(getBoroughsForSpecificCrime(specific_crime),'green',10,10,'Per Capita Knife Crimes','Knife Crimes Committed','London Boroughs')
```

#### 2.10.3 Selecting top 10% boroughs with lowest crimes

To make it easier for user to choose borough, data mentioned above can be further filtered down show only top 10% of London boroughs with the lowest number of specific (e.g. Knife) crimes as shown below:

```python
#Top 10% of all London boroughs with lowest knife crime
df_sorted = getBoroughsForSpecificCrime(specific_crime)
df_sorted = df_sorted.head(int(len(df_sorted)*(10/100)))
drawBarGraph(df_sorted,'green',5,2,'Recommended Boroughs','Knife Crimes','London Boroughs')
```

```python
#########################################################
#Show trend of multiple crimes in specific London borough
#########################################################

#Select months from columns of DataFrame to show on X-axis
x_axis = crime_df.columns.values[3:27]
#Set graph size
plt.figure(figsize=(20, 20))
#Select row containing crime named 'Arson' and show it on Y-axis
y_axis = resultDf.iloc[0][3:27]
#Set x-axis and y-axis and label 'Arson'
plt.plot(x_axis,y_axis,"--ro", label = "Arson")
#Select row containing crime named 'Criminal Damage' and show it on Y-axis
y_axis = resultDf.iloc[1][3:27]
#Set x-axis and y-axis and label 'Criminal Damage'
plt.plot(x_axis,y_axis,"--bo", label = "Criminal Damage")
#Select row containing crime named 'Burglary Business and Community' and show it on Y-axis
y_axis = resultDf.iloc[2][3:27]
#Set x-axis and y-axis and label 'Burglary Business and Community'
plt.plot(x_axis,y_axis,"--go", label = "Burglary Business and Community")
#Select row containing crime named 'Domestic Burglary' and show it on Y-axis
y_axis = resultDf.iloc[3][3:27]
#Set x-axis and y-axis and label 'Domestic Burglary'
plt.plot(x_axis,y_axis,"--yo", label = "Domestic Burglary")
#Select row containing crime named 'Drug Trafficking' and show it on Y-axis
y_axis = resultDf.iloc[4][3:27]
#Set x-axis and y-axis and label 'Drug Trafficking'
plt.plot(x_axis,y_axis,"--co", label = "Drug Trafficking")
#Select row containing crime named 'Possession of Drugs' and show it on Y-axis
y_axis = resultDf.iloc[5][3:27]
#Set x-axis and y-axis and label 'Possession of Drugs'
plt.plot(x_axis,y_axis,"--mo", label = "Possession of Drugs")
#Select row containing crime named 'Dangerous Driving' and show it on Y-axis
y_axis = resultDf.iloc[6][3:27]
#Set x-axis and y-axis and label 'Dangerous Driving'
plt.plot(x_axis,y_axis,"--ko", label = "Dangerous Driving")
#Set title of graph
plt.title("Crime Trend in Hounslow", fontsize = 20)
#Set label of X-axis
plt.xlabel("Time Period", fontsize = 20)
#Set label of Y-axis
plt.ylabel("Number of crimes", fontsize = 20)
#Set graph lagend
plt.legend(prop={"size":16})
#Show graph
plt.show()
```

#### 2.10.4 Comparing all crimes in specific borough

Finally, users need to view all crimes in their selected borough over a period of time (2 years) so they get complete picture before making final decision. At this point, user can go (zoom out) to higher level and check again all boroughs for specific crime if they want to review their choice of borough and choose a completely different borough, again looking overall crimes in that borough. This back and forth is natural part of human decision making before settling for final decision which makes the proposed approach more useful.  

```python
#########################################################
#show all crimes in specific london borough
#########################################################

#Select all the crimes and sum/add for all the months
x_axis = list(resultDf.iloc[:,3:27].sum(axis='columns'))
#Select list of all crimes
y_axis = list(resultDf["MinorText"])
#Create new DataFrame so it can be sorted and used for plotting
dfTotalCrimesInBorough = pd.DataFrame(
    dict(
        namesOfCrimes = y_axis,
        numberOfcrimes = x_axis
    
    )
)
#Sort DataFrame
df_sorted = dfTotalCrimesInBorough.sort_values("numberOfcrimes")
#Set graph size
plt.figure(figsize=(10, 10))
#Set sorted DataFrame and other parameters to draw bar graph
plt.barh("namesOfCrimes","numberOfcrimes", data = df_sorted, color = "blue")
#Set title of graph
plt.title('Criminal activity in Hounslow', fontsize = 17)
#Set label for X-axis
plt.xlabel('Number of Crimes Committed', fontsize = 15)
#Set label for Y-axis
plt.ylabel('Name of Crimes', fontsize = 15)
#Show graph
plt.show()
```

Word cloud is a powerful visualisation tool that is used to make easier for user to see that which category of crime is more prevalent in the borough. For example word cloud of Hounslow shows that crimes under the category of 'Violence without Injury' is significant in this London borough.

Word cloud has been generated based on the frequency of category of crimes. To achieve it, a dataframe is created based on crime names and their frequencies. Dataframe is coverted to dictionary and that is used to generate world cloud.

```python
crime_names = list(resultDf["MinorText"])
#Since each category is made of many words so for wordcloud to treat whole category as one word replace space between words with underscore
crime_names = np.char.replace(crime_names, " ", "_")
#Get sum of crimes for each category
crime_sums = list(resultDf.iloc[:, 3:27].sum(axis='columns'))
#Create dataframe from crimes name (category) and number of crimes for each category
df_text = pd.DataFrame(
    dict(
        names=crime_names,
        crime_num=crime_sums
    )
)
#Convert dataframe to dictionary so it can be used by world cloud
cloud_data = df_text.set_index('names').to_dict()['crime_num']

# Create and generate a word cloud image based on the frequency of each category of crime:
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate_from_frequencies(cloud_data)

# Display the generated image:
plt.figure(figsize=(40, 40))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
